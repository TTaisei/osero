# 実験1

```python
import pandas as pd

df = pd.read_csv("data.csv")
```

```python
len(df)
```

```python
df.head()
```

```python
df_sorted = df.sort_values(by="test_MAE", inplace=False)
df_sorted.head(20)
```

```python
optimizers = df["optimizer"].unique()
df_top = df.sort_values(by="test_MSE", inplace=False).head(25)
df_worst = df.sort_values(by="test_MSE", inplace=False, ascending=False).head(25)

print("top")
for opt in optimizers:
    num = 0
    num += len(df_top.query("optimizer=='%s'" % opt))

    print("%s:\t%d" % (opt, num))

print("\nworst")
for opt in optimizers:
    num = 0
    num += len(df_worst.query("optimizer=='%s'" % opt))

    print("%s:\t%d" % (opt, num))
```

```python
hook_functions = df["hook_function"].unique()

print("top")
for hf in hook_functions:
    num = 0
    num += len(df_top.query("hook_function=='%s'" % hf))

    print("%s:\t%d" % (hf, num))

print("\nworst")
for hf in hook_functions:
    num = 0
    num += len(df_worst.query("hook_function=='%s'" % hf))

    print("%s:\t%d" % (hf, num))
```

# 実験2

```python
import pandas as pd

df = pd.read_csv("data2.csv")
```

```python
df.head()
```

```python
df_sorted = df.sort_values(by="test_MAE", inplace=False)
df_sorted.head(20)
```

```python
batich_sizes = df["batch_size"].unique()
df_top = df.sort_values(by="test_MSE", inplace=False).head(25)
df_worst = df.sort_values(by="test_MSE", inplace=False, ascending=False).head(25)

print("batch_size:")
print("top")
for bs in batich_sizes:
    num = 0
    num += len(df_top.query("batch_size==%d" % bs))

    print("%d:\t%d" % (bs, num))

print("\nworst")
for bs in batich_sizes:
    num = 0
    num += len(df_worst.query("batch_size==%d" % bs))

    print("%d:\t%d" % (bs, num))
```

```python
optimizers = df["optimizer"].unique()

print("optimizer: ")
print("top")
for opt in optimizers:
    num = 0
    num += len(df_top.query("optimizer=='%s'" % opt))

    print("%s:\t%d" % (opt, num))

print("\nworst")
for opt in optimizers:
    num = 0
    num += len(df_worst.query("optimizer=='%s'" % opt))

    print("%s:\t%d" % (opt, num))
```

```python
hook_functions = df["hook_function"].unique()

print("hook_function:")
print("top")
for hf in hook_functions:
    num = 0
    num += len(df_top.query("hook_function=='%s'" % hf))

    print("%s:\t%d" % (hf, num))

print("\nworst")
for hf in hook_functions:
    num = 0
    num += len(df_worst.query("hook_function=='%s'" % hf))

    print("%s:\t%d" % (hf, num))
```
